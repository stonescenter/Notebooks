```python
# funçoes para gerar o dataset
def get_data(ds):
  municipios = ds['CODUFMUN'].unique()
  leitos = []
  for m in municipios:
    leito = {'municipio' : '', 
          'leitos_sus': [], 
          'leitos_nosus': []
    }

    leito['municipio'] = m

    for r in ds.iterrows():
      compe = str(r[0])
      ano, mes = compe[:4], compe[4:]
      municipio = r[1]['CODUFMUN']
      sus, nosus = r[1]['QT_SUS'], r[1]['QT_NSUS']

      #print(ano, mes, municipio, sus, nosus)
      if str(m) == str(municipio):
        leito['leitos_sus'].append(sus)
        leito['leitos_nosus'].append(nosus)

    size1 = len(leito['leitos_sus'])
    size2 = len(leito['leitos_nosus']) 

    if size1 != 12:
      leito['leitos_sus'].extend([0]*(12-size1))


    if size2 != 12:
      leito['leitos_nosus'].extend([0]*(12-size2))

    leitos.append(leito)

  return leitos

def get_data_by_municipio(lst : list, m: int):
  
  for l in lst:
    if l['municipio'] == m:
      return l
```

```python
# construimos nosso dataset por ano
data19 = get_data(qsus19)
data20 = get_data(qsus20)
data21 = get_data(qsus21)
```

```python
import plotly.graph_objects as go
import plotly.figure_factory as ff
import numpy as np
from operator import itemgetter

def plot_ds(data, municipio, year):

  # Initialize a figure with ff.create_table(table_data)
  #fig = ff.create_table(table_data, height_constant=60)
  fig = go.Figure()

  res = get_data_by_municipio(data, municipio )


  meses = ['Jan', 'Feb', 'Mar', 'Abr', 'Jun', 'Jul', 'Ago', 'Set', 'Out', 'Nov', 'Dec']
  # Make traces for graph
  trace1 = go.Bar(x=meses, y=res['leitos_sus'], xaxis='x2', yaxis='y2',
                  marker=dict(color='#0099ff'),
                  name='Leitos SUS')
  trace2 = go.Bar(x=meses, y=res['leitos_nosus'], xaxis='x2', yaxis='y2',
                  marker=dict(color='#404040'),
                  name='Leitos No SUS')

  # Add trace data to figure
  fig.add_traces([trace1, trace2])

  # initialize xaxis2 and yaxis2
  fig['layout']['xaxis2'] = {}
  fig['layout']['yaxis2'] = {}

  # Edit layout for subplots
  #fig.layout.yaxis.update({'domain': [0,  np.max(sus_list)]})
  #fig.layout.yaxis2.update({'domain': [0,  np.max(nosus_list)]})
  fig.layout.yaxis.update({'domain': [0, .45]})
  fig.layout.yaxis2.update({'domain': [.6, 1]})


  # The graph's yaxis2 MUST BE anchored to the graph's xaxis2 and vice versa
  fig.layout.yaxis2.update({'anchor': 'x2'})
  fig.layout.xaxis2.update({'anchor': 'y2'})
  fig.layout.yaxis2.update({'title': 'Quantidade'})

  # Update the margins to add a title and see graph x-labels.
  #fig.layout.margin.update({'t':75, 'l':50})
  fig.layout.update({'title': 'Municipio:' + str(municipio) + " <br>Ano:" + str(year)})

  # Update the height because adding a graph vertically will interact with
  # the plot height calculated for the table
  fig.layout.update({'width':900})

  # Plot!
  fig.show()
```

```python
# Calcula o numero de leitos e no leitos para um determinado municipio por ano e vizualiza.
muni = 520680
plot_ds(data19,muni, 2019)
plot_ds(data20,muni, 2020)
plot_ds(data21,muni, 2021)
```

```python
import plotly.graph_objects as go

def plot_simple(data):
  municipio_list = list(map(itemgetter('municipio'), data ))
  municipio_list = [str(m) for m in municipio_list]
  sus_list = list(map(itemgetter('leitos_sus'), data ))
  nosus_list = list(map(itemgetter('leitos_nosus'), data ))

  fig = go.Figure()
  fig.add_trace(go.Histogram(
        histfunc="count", 
        #histnorm='percent',
        y=sus_list, 
        x=municipio_list,  
        xbins=dict( # bins used for histogram
          start=0,
          end=len(municipio_list)),
          autobinx=True,
          name="Sus")
        )
  
  fig.add_trace(go.Histogram(
        histfunc="sum",
        #histnorm='percent',
        y=nosus_list, 
        x=municipio_list,  
        xbins=dict( # bins used for histogram
          start=0,
          end=len(municipio_list)),
        name="No Sus")
        )

  fig.show()
```
